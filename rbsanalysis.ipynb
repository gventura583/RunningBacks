```python
import pandas as pd
rbsdf = pd.read_csv('rbs.csv')
rbsdf
```

```python
rbsdf['Player'].value_counts()
```

- Check values in the 'Player' column and notice that there's special characters

```python
rbsdf['Player'] = rbsdf['Player'].str.replace('*+', '', regex=False)
rbsdf['Player'] = rbsdf['Player'].str.replace('*', '', regex=False)
rbsdf['Player'].value_counts()
```

- Cleaned the '*' and '+' characters from the player's names

```python
stats2021 = rbsdf[rbsdf['Year'] == 2021].sort_values('Att', ascending=False).head(25)
stats2021
```

- Created a new data fram with only stats from the year 2021

```python
stats2021.plot.bar('Player', 'Att')
```

```python
stats2021.plot.bar('Player', 'Yds')
```

```python
stats2021.plot.bar('Att', 'Yds')
```

```python
stats2021.plot.bar('Player', 'Y/A')
```

```python
stats2021.corr()['Yds']
```
